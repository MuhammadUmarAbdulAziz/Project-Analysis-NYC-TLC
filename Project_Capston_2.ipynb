## Data

*Penjelasan Kolom*

VendorID - ID penyedia layanan taksi. Biasanya:

- 1 = Creative Mobile Technologies (CMT)
- 2 = VeriFone Inc.

lpep_pickup_datetime - Waktu penjemputan penumpang.

lpep_dropoff_datetime - Waktu penurunan penumpang.

store_and_fwd_flag - Indikator apakah data perjalanan sementara disimpan sebelum dikirim ke server:

- Y = Data disimpan sementara.
- N = Data dikirim langsung ke server.

RatecodeID - Kode tarif yang digunakan dalam perjalanan:

- 1 = Tarif standar.
- 2 = Tarif perjalanan ke bandara JFK.
- 3 = Tarif perjalanan ke bandara Newark.
- 4 = Perjalanan dengan tarif yang dinegosiasikan sebelumnya.
- 5 = Tarif perjalanan gratis.
- 6 = Perjalanan ke luar kota yang telah ditentukan tarifnya.

PULocationID - ID lokasi penjemputan (berdasarkan zona taksi NYC).

DOLocationID - ID lokasi tujuan penumpang (berdasarkan zona taksi NYC).

passenger_count - Jumlah penumpang dalam perjalanan.

trip_distance - Jarak perjalanan dalam mil.

fare_amount - Tarif dasar perjalanan dalam dolar.

extra - Biaya tambahan, seperti:

- 1.00 = Tambahan biaya perjalanan malam hari (20:00 - 06:00).
- 0.50 = Biaya tambahan jam sibuk.

mta_tax - Pajak Metro Transit Authority (MTA) sebesar $0.50 per perjalanan.

tip_amount - Jumlah tip yang diberikan penumpang (jika pembayaran non-tunai).

tolls_amount - Jumlah biaya tol selama perjalanan.

ehail_fee - Biaya pemesanan elektronik (biasanya NaN untuk taksi reguler).

improvement_surcharge - Biaya tambahan per perjalanan ($0.30).

total_amount - Total biaya perjalanan termasuk tarif, pajak, tol, dan tip.

payment_type - Jenis pembayaran:

- 1 = Kartu kredit.
- 2 = Tunai.
- 3 = Tanpa biaya (misalnya perjalanan promosi).
- 4 = Dispute (pembayaran dipermasalahkan).
- 5 = Pembayaran lain.
- 6 = Voided trip (perjalanan yang dibatalkan).

trip_type - Jenis perjalanan:

- 1 = Perjalanan reguler.
- 2 = Perjalanan yang dipesan sebelumnya (dispatch).

congestion_surcharge - Biaya kemacetan ($2.75 per perjalanan, berlaku untuk perjalanan di Manhattan).

### 1. Menghapus Data Tidak Logis

Beberapa kondisi data yang dianggap tidak valid dan perlu dihapus dari dataset antara lain:
- `Passenger_count` bernilai 0
- `Trip_distance` bernilai kurang dari atau sama dengan 0
- `Total_amount` bernilai negatif

Data dengan karakteristik seperti ini dianggap tidak masuk akal dalam konteks perjalanan taksi.


```python
# Hapus data dengan penumpang <= 0
df = df[df['passenger_count'] > 0]

# Hapus data dengan jarak perjalanan <= 0
df = df[df['trip_distance'] > 0]

# Hapus data dengan total_amount negatif
df = df[df['total_amount'] >= 0]

print("Data setelah pembersihan nilai tidak logis:", df.shape)
```

### 2. Mengubah Format Waktu

Kolom `lpep_pickup_datetime` dan `lpep_dropoff_datetime` perlu diubah menjadi format `datetime` agar bisa digunakan untuk analisis waktu seperti durasi perjalanan, jam sibuk, dll.

```python
# Konversi kolom waktu
df['lpep_pickup_datetime'] = pd.to_datetime(df['lpep_pickup_datetime'], errors='coerce')
df['lpep_dropoff_datetime'] = pd.to_datetime(df['lpep_dropoff_datetime'], errors='coerce')

# Cek apakah ada nilai yang gagal dikonversi
print("Jumlah NaT setelah konversi pickup datetime:", df['lpep_pickup_datetime'].isna().sum())
print("Jumlah NaT setelah konversi dropoff datetime:", df['lpep_dropoff_datetime'].isna().sum())

# Drop baris yang gagal dikonversi datetime
df = df.dropna(subset=['lpep_pickup_datetime', 'lpep_dropoff_datetime'])
```

### 3. Validasi Waktu Perjalanan

Pastikan waktu drop-off selalu lebih besar dari waktu pickup. Jika tidak, data tersebut dianggap tidak valid.


```python
# Validasi: Dropoff harus sesudah pickup
df = df[df['lpep_dropoff_datetime'] > df['lpep_pickup_datetime']]

print("Data setelah validasi waktu:", df.shape)
```

### 4. Hapus Kolom yang Tidak Relevan

Jika ada kolom yang seluruhnya kosong atau tidak digunakan dalam analisis, bisa dihapus untuk efisiensi.

# DATA ANALYSIS

### 1. Analisis Jumlah Perjalanan per Hari (Tren Aktivitas)
- Tujuan: Mengetahui hari-hari dengan aktivitas tertinggi atau terendah dalam penggunaan layanan taksi.

```python
df_merged['pickup_day'] = df_merged['lpep_pickup_datetime'].dt.day_name()

trips_per_day = df_merged['pickup_day'].value_counts().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)

plt.figure(figsize=(10, 5))
sns.barplot(x=trips_per_day.index, y=trips_per_day.values, palette="Blues_d")
plt.title('Jumlah Perjalanan per Hari')
plt.ylabel('Jumlah Perjalanan')
plt.xlabel('Hari')
plt.xticks(rotation=45)
plt.grid(axis='y')
for i, value in enumerate(trips_per_day.values):
    plt.text(i, value + max(trips_per_day.values) * 0.01,
             f'{value}', ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

INSIGHT:

Mayoritas perjalanan dilakukan pada hari kerja (terutama hari Selasa), menunjukkan aktivitas penumpang tinggi menjelang akhir pekan. Hari Minggu cenderung paling sepi.

- Distribusi per jam dalam sehari → Untuk mengetahui jam sibuk (peak hours) dan potensi efisiensi armada.

```python
df_merged['pickup_hour'] = df_merged['lpep_pickup_datetime'].dt.hour
hourly_distribution = df_merged['pickup_hour'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x=hourly_distribution.index, y=hourly_distribution.values, marker='o', color='green')
plt.title('Distribusi Perjalanan per Jam', fontsize=14)
plt.xlabel('Jam (0-23)', fontsize=12)
plt.ylabel('Jumlah Perjalanan', fontsize=12)
plt.xticks(range(0, 24))
plt.grid(True)
for x, y in zip(hourly_distribution.index, hourly_distribution.values):
    plt.text(x, y + max(hourly_distribution.values) * 0.01,
             f'{y}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

Insight:

Jam pagi (7–9) dan sore (17–19) menunjukkan peningkatan signifikan — cocok untuk strategi insentif pengemudi saat peak time.

Acttionable Steps:

- Penyesuaian Armada: Tambahkan armada aktif di waktu-waktu sibuk (misal 7–9 pagi, 5–7 sore).

- Insentif Pengemudi: Berikan bonus khusus untuk shift pagi/sore agar lebih banyak pengemudi tersedia saat jam sibuk.

- Perencanaan Shift: Jadwalkan shift berdasarkan tren aktivitas harian untuk menghindari kelebihan/kekurangan kendaraan.

### 2. Analisis Metode Pembayaran

- Tujuan: Mengetahui preferensi pembayaran oleh penumpang.

```python
payment_map = {
    1: 'Credit Card', 2: 'Cash', 3: 'No Charge',
    4: 'Dispute', 5: 'Unknown', 6: 'Voided Trip'
}
df_merged['payment_label'] = df_merged['payment_type'].map(payment_map)

payment_counts = df_merged['payment_label'].value_counts().reset_index()
payment_counts.columns = ['Payment Type', 'Count']

plt.figure(figsize=(10, 6))
ax = sns.barplot(data=payment_counts, x='Payment Type', y='Count', palette='Set2')
plt.title('Distribusi Jenis Pembayaran', fontsize=14)
plt.xlabel('Jenis Pembayaran', fontsize=12)
plt.ylabel('Jumlah Perjalanan', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}',
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom',
                fontsize=9, fontweight='bold',
                xytext=(0, 5), textcoords='offset points')
plt.tight_layout()
plt.show()
```

Insight:

Lebih dari 70% penumpang membayar dengan kartu kredit, mengindikasikan pentingnya menjaga sistem pembayaran digital yang lancar dan aman.

- Tip berdasarkan metode pembayaran → Apakah pengguna kartu kredit memberi tip lebih besar?

```python
avg_tip_by_payment = df_merged.groupby('payment_label')['tip_amount'].mean().sort_values(ascending=False)
ax = avg_tip_by_payment.plot(kind='bar', color='orange', figsize=(8, 5))
plt.title("Rata-rata Tip Rate Berdasarkan Metode Pembayaran")
plt.ylabel("Tip Rate")
plt.grid(axis='y')
for i, value in enumerate(avg_tip_by_payment):
    ax.text(i, value + max(avg_tip_by_payment) * 0.01,  # sedikit di atas batang
            f'{value:.2f}', ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.show()
```

 Insight:

Penumpang yang menggunakan kartu kredit memberikan tip lebih tinggi dibandingkan cash — mendukung perluasan sistem pembayaran nontunai.

Actionable Steps:

- Dorong Pembayaran Digital: Berikan potongan harga kecil untuk pembayaran non-tunai.

- Kemitraan Fintech: Jalin kerja sama dengan penyedia e-wallet untuk cashback atau rewards khusus pengguna transportasi TLC.

- Kampanye Edukasi: Beri edukasi ke pengemudi untuk mendorong penumpang membayar secara digital (lebih aman dan transparan).

### 3. Analisis Lokasi Terpopuler (Pickup dan Dropoff)
- Tujuan: Mengidentifikasi area dengan permintaan tertinggi untuk optimasi lokasi armada.

```python
location_zone_map = dict(zip(df_merged['PULocationID'], df_merged['Zone']))
df_merged['PickupZone'] = df_merged['PULocationID'].map(location_zone_map)
df_merged['DropoffZone'] = df_merged['DOLocationID'].map(location_zone_map)

top_pickup = df_merged['PickupZone'].value_counts().head(10)
top_dropoff = df_merged['DropoffZone'].value_counts().head(10)

top_pickup = top_pickup.sort_values(ascending=False)
top_dropoff = top_dropoff.sort_values(ascending=False)


fig, axs = plt.subplots(1, 2, figsize=(14, 6))

sns.barplot(x=top_pickup.index, y=top_pickup.values, ax=axs[0], palette="viridis")
axs[0].set_title("Top 10 Pickup Zones (Descending)", fontsize=14)
axs[0].set_xlabel("Pickup Zone", fontsize=12)
axs[0].set_ylabel("Jumlah", fontsize=12)
axs[0].tick_params(axis='x', rotation=45)
for i, value in enumerate(top_pickup.values):
    axs[0].text(i, value + 0.5, str(value), ha='center', fontsize=10, color='black')

sns.barplot(x=top_dropoff.index, y=top_dropoff.values, ax=axs[1], palette="magma")
axs[1].set_title("Top 10 Dropoff Zones (Descending)", fontsize=14)
axs[1].set_xlabel("Dropoff Zone", fontsize=12)
axs[1].set_ylabel("Jumlah", fontsize=12)
axs[1].tick_params(axis='x', rotation=45)
for i, value in enumerate(top_dropoff.values):
    axs[1].text(i, value + 0.5, str(value), ha='center', fontsize=10, color='black')
plt.tight_layout()
plt.show()
```

Insight:

Lokasi-lokasi seperti Midtown dan JFK Airport sering muncul sebagai titik pickup dan dropoff utama — bisa menjadi fokus untuk pemasaran atau penambahan armada.

- Durasi perjalanan antar lokasi populer → Bisa digunakan untuk menentukan efisiensi rute atau harga dinamis.

```python
from itertools import product

df_merged['trip_duration_minutes'] = (df_merged['lpep_dropoff_datetime'] - df_merged['lpep_pickup_datetime']).dt.total_seconds() / 60

# Bersihkan durasi yang tidak masuk akal (misal negatif atau lebih dari 180 menit)
df_merged = df_merged[(df_merged['trip_duration_minutes'] > 0) & (df_merged['trip_duration_minutes'] <= 180)]

# Tentukan top lokasi pickup paling banyak
top_pickup = df_merged['PULocationID'].value_counts().head(10)

# Buat pasangan lokasi populer
popular_locs = top_pickup.index.tolist()
pairs = list(product(popular_locs, popular_locs))

# Filter data hanya yang termasuk top pickup & dropoff
durasi_lokasi = df_merged[
    df_merged['PULocationID'].isin(popular_locs) & 
    df_merged['DOLocationID'].isin(popular_locs)
]

# Hitung rata-rata durasi
mean_durasi = durasi_lokasi.groupby(['PULocationID', 'DOLocationID'])['trip_duration_minutes'].mean().unstack()

# Visualisasi dengan heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(mean_durasi, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Durasi Rata-rata antar Lokasi Populer (menit)")
plt.xlabel("Dropoff Location ID")
plt.ylabel("Pickup Location ID")
plt.tight_layout()
plt.show()
```

Insight:

Beberapa rute antar lokasi populer menunjukkan durasi tinggi. Misalnya, rute JFK ke Midtown bisa digunakan untuk pengaturan tarif dinamis atau jalur cepat (express).

Acrionable Steps:
- Optimasi Rute: Sediakan rute khusus atau ekspres dari titik-titik populer ke lokasi tujuan utama (misal, JFK ke Midtown).

- Penempatan Zona Khusus: Buat zona tunggu khusus untuk taksi di titik pickup favorit agar antrian lebih tertib.

- Pricing Strategy: Terapkan tarif dinamis (dynamic pricing) untuk area dengan permintaan tinggi.

### 4. Analisis Tip Rate (Tingkat Kedermawanan)
- Tujuan: Mengetahui area atau kondisi di mana penumpang memberi tip lebih besar.

```python
df['tip_rate'] = df['tip_amount'] / df['total_amount']
df = df[(df['tip_rate'].notna()) & (df['tip_rate'] < 1)]

plt.figure(figsize=(10, 5))
sns.histplot(df['tip_rate'], bins=50, kde=True, color='green')
plt.title('Distribusi Tip Rate (Tip / Total Amount)', fontsize=14)
plt.xlabel('Tip Rate (%)', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.xlim(0, 0.5)
plt.grid(True)
counts, edges, patches = plt.hist(df['tip_rate'], bins=50, range=(0, 0.5), color='green', alpha=0)
for count, patch in zip(counts, patches):
    if count > 0:
        x = patch.get_x() + patch.get_width() / 2
        y = patch.get_height()
        plt.text(x, y + 0.5, f'{int(count)}', ha='center', fontsize=8, color='black')
plt.tight_layout()
plt.show()
```

Insight:

Mayoritas penumpang memberikan tip antara 10–20% dari total pembayaran, dan tip tertinggi lebih banyak muncul di perjalanan berjarak menengah dan menggunakan kartu kredit.

- Analisis lokasi dengan tip rate tertinggi → Mengetahui area di mana penumpang paling “loyal”.

```python
df_merged['tip_rate'] = df_merged['tip_amount'] / df_merged['total_amount']
df_merged = df_merged[(df_merged['tip_rate'].notna()) & (df_merged['tip_rate'] < 1)]
location_zone_map = dict(zip(df_merged['PULocationID'], df_merged['Zone']))
df_merged['PickupZone'] = df_merged['PULocationID'].map(location_zone_map)
tip_by_area = df_merged.groupby('PickupZone')['tip_rate'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
barplot = sns.barplot(x=tip_by_area.index, y=tip_by_area.values, palette="Greens")
plt.title("10 Lokasi Pickup (Zone) dengan Tip Rate Tertinggi", fontsize=14)
plt.xlabel("Zona Pickup", fontsize=12)
plt.ylabel("Rata-rata Tip Rate (%)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
for index, value in enumerate(tip_by_area.values):
    barplot.text(index,value + 0.01,f"{value:.2%}",ha='center',fontsize=9,color='black')
plt.tight_layout()
plt.show()
```

dibeberapa lokasi komersial memberikan tip tertinggi. Dapat dijadikan acuan untuk peningkatan layanan atau insentif pengemudi di area tersebut.
